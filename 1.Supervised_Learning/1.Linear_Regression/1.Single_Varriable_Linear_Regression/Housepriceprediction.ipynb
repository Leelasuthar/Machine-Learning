```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./house-prices.csv")
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['SqFt']],df.Price)
```

```python
reg.predict([[5000]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.title('home price prediction')
plt.xlabel('SqFt')
plt.ylabel('Price')
plt.scatter(df.SqFt, df.Price ,marker='*',color='purple')
```
