```python
import pandas as pd
import  numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv(".\house-prices.csv")
df
```

```python
import math
median_Bedrooms = math.floor(df.Bedrooms.median())
median_Bedrooms
```

```python
#Handling/filling the missing values 
df.Bedrooms = df.Bedrooms.fillna(median_Bedrooms)
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['SqFt','Bedrooms','Bathrooms','Offers']],df.Price)
```

```python
reg.coef_
```

```python
reg.intercept_
```
